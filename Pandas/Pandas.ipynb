```python
# Imported the necessary libraries for project.

import pandas as pd

import numpy as np

import datetime as dt

from pathlib import Path

%matplotlib inline
```

```python
# Created a path file for algo_returns.csv

algo_returns_path = Path("Resources/algo_returns.csv")

# Created a variable named "whale_returns_df" and converted the dates to a DateTimeIndex.

algo_returns_df = pd.read_csv(algo_returns_path, index_col ='Date', infer_datetime_format = True, parse_dates = True)

algo_returns_df.sort_index(ascending=True)
```

```python
# Used the `dtypes` attribute to list the column of data types from algo_returns_df

# Dtype float means there is no "$" sign or characters found

algo_returns_df.dtypes
```

```python
# Assessed data quality by determining the number of nulls from algo_returns_df

algo_returns_df.isnull().sum()
```

```python
# Dropped records with null from algo_returns_df

algo_returns_df.dropna(inplace = True)

# Initialized algo_returns_df

algo_returns_df.sort_index(ascending=True)
```

```python
# Created a path file for # Created a path file for sp_tsx_history_path.csv

sp_tsx_history_path = Path("Resources/sp_tsx_history.csv")

# Created a variable named "_df" and converted the dates to a DateTimeIndex.

sp_tsx_history_df = pd.read_csv(sp_tsx_history_path, index_col = 'Date', infer_datetime_format = True, parse_dates = True)

sp_tsx_history_df.sort_index(ascending=True)
```

```python
# Used the `dtypes` attribute to list the column of data types from sp_tsx_history_df

sp_tsx_history_df.dtypes
```

```python
# Assessed data quality by checking nulls

sp_tsx_history_df.isnull()
```

```python
# Assessed data quality by determining the number of nulls from sp_tsx_history_df

sp_tsx_history_df.isnull().sum()
```

```python
# Dropped records with null from sp_tsx_history_df

sp_tsx_history_df.dropna(inplace = True)
```

```python
# Removed "$" character from sp_tsx_history_df 

sp_tsx_history_df["Close"] = sp_tsx_history_df["Close"].replace({"\$":""}, regex = True)

sp_tsx_history_df.sort_index(ascending=True)
```

```python
# Removed "," character from sp_tsx_history_df 

sp_tsx_history_df["Close"] = sp_tsx_history_df["Close"].replace({",":""}, regex = True)

sp_tsx_history_df.sort_index(ascending=True)
```

```python
# Fixed Data Types for sp_tsx_history_df to float "numeric"

sp_tsx_history_df["Close"] = sp_tsx_history_df["Close"].astype(float)

# Rechecked dtypes for sp_tsx_history_df

sp_tsx_history_df.dtypes

# Initialized sp_tsx_history_df

sp_tsx_history_df.sort_index(ascending=True)
```

```python
# Calculated the S&P 60 daily return

sp_tsx_returns_df = sp_tsx_history_df.pct_change()

# Intialized sp_tsx_returns

sp_tsx_returns_df.sort_index(ascending=True)
```

```python
# Assessed data quality by determining the number of nulls from sp_tsx_history_df

sp_tsx_returns_df.isnull().sum()
```

```python
# Dropped records with null from sp_tsx_history_df

sp_tsx_returns_df.dropna(inplace = True)

# Initialized sp_tsx_returns

sp_tsx_returns_df.sort_index(ascending=True)
```

```python
# Renamed "Close" column from sp_tsx_returns to "S&P TSX 60"

sp_tsx_returns_df = sp_tsx_returns_df.rename(columns = {"Close":"S&P TSX 60"})

# Initialized sp_tsx_returns

sp_tsx_returns_df.sort_index(ascending=True)
```

```python
# Combined Whale, Algorithmic, and S&P TSX 60 Returns

combined_df = pd.concat([whale_returns_df, algo_returns_df, sp_tsx_returns_df], axis="columns", join="inner")

# Initialized combined_df

combined_df.sort_index(ascending=True)
```

```python
# Conducted Quantitative Analysis

# Performance Analysis: Plotted daily returns

combined_df.plot(figsize=(20,10), title = "Daily Returns")
```

```python
# Perfornmance Analysis: Calculated and Plotted cumulative returns

cumulative_returns_df = (1 + combined_df).cumprod()

cumulative_returns_df.plot(figsize=(20,10))
```

```python
# Risk Analysis : Created a box plot for each of the returns

combined_boxplot_df = combined_df.boxplot(figsize = (20,10), grid = False)
```

```python
# Risk Analysis : Calculated the standard deviation of each portfolio

combined_std_df = combined_df.std()

combined_std_df.sort_values(ascending=False, inplace=True)

combined_std_df
```

```python
# Risk Analysis: Determined which portfolios are riskier than the S&P TSX 60

combined_std_df > combined_std_df["S&P TSX 60"]
```

```python
# Risk Analysis: Calculated the annualized standard deviation (252 trading days)

annual_std_df = combined_df.std() * np.sqrt(252)

annual_std_df.sort_values(ascending=False)
```

```python
# Rolling Statistics: Calculated and plot the rolling standard deviation for all portfolios using a 21-day window.

group_rolling_df = combined_df.rolling(window=21).std().plot(figsize = (20,10))
```

```python
# Rolling Statistics: Calculated and plotted the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.

import seaborn as sns

correlation_df = combined_df.corr()

correlation_df

sns.heatmap(correlation_df, vmin = -1, vmax = 1, annot=True, cmap="BuPu") 
```

```python
# Calculated and plotted Beta for a chosen portfolio and the S&P 60 TSX

rolling_covariance = combined_df["TIGER GLOBAL MANAGEMENT LLC"].rolling(window = 60).cov(combined_df["S&P TSX 60"])

rolling_variance = combined_df["S&P TSX 60"].rolling(window = 60).var()

rolling_combined_beta_df = rolling_covariance / rolling_variance

rolling_combined_beta_df.plot(figsize = (20,10)) 
```

```python
# Rolling Statistics Challenge: Calculated Exponentially Weighted Average

ewm = combined_df.ewm(min_periods = 21, halflife = 21).mean().plot(figsize=(20,10))
```

```python
# Sharpe Ratios: Calculated Annualized Sharpe Ratios

sharpe_ratios_df = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))

sharpe_ratios_df
```

```python
# Visualized the sharpe_ratios_df as a bar plot

sharpe_ratios_df.plot(kind = "bar")
```

```python
# Created Custom Portfolio: Shopify

# Created a path file for shop_historical.csv

shop_historical_path = Path("Resources/shop_historical.csv")

# Created a variable named "shopify_df" and converted the dates to a DateTimeIndex.

shopify_df = pd.read_csv(shop_historical_path, index_col ='Date', infer_datetime_format = True, parse_dates = True)

# Changed and renamed columns to better reflect data

shopify_df.drop(columns=["Symbol"], inplace = True)

shopify_df.rename(columns = {"Close":"SHOP"}, inplace=True)

shopify_df.sort_index(ascending=True)
```

```python
# Created Custom Portfolio: Open Text Corporation

# Created a path file for otex_historical.csv

otex_historical_path = Path("Resources/otex_historical.csv")

# Created a variable named "otex_df" and converted the dates to a DateTimeIndex.

otex_df = pd.read_csv(otex_historical_path, index_col ='Date', infer_datetime_format = True, parse_dates = True)

# Changed and renamed columns to better reflect data

otex_df.drop(columns=["Symbol"], inplace = True)

otex_df.rename(columns = {"Close":"OTEX"}, inplace=True)

otex_df.sort_index(ascending=True)
```

```python
# Created Custom Portfolio: Loblaw Companies Limited

# Created a path file for otex_historical.csv

l_historical_path = Path("Resources/l_historical.csv")

# Created a variable named "loblaw_df" and converted the dates to a DateTimeIndex.

loblaw_df = pd.read_csv(l_historical_path, index_col ='Date', infer_datetime_format = True, parse_dates = True)

# Changed and renamed columns to better reflect data

loblaw_df.drop(columns=["Symbol"], inplace = True)

loblaw_df.rename(columns = {"Close":"LOBLAW"}, inplace=True)

loblaw_df.sort_index(ascending=True)
```

```python
# Custom Portforlio: Concatenated all stocks into a single DataFrame

combined_custom_df = pd.concat([shopify_df, otex_df, loblaw_df], axis = 'columns', join = 'inner' )

# Fixed index

combined_custom_df.sort_index(ascending=True)
```

```python
# Custom Portfolio: Dropped Nulls

combined_custom_df.dropna()

combined_custom_df.isnull().sum()
```

```python
# Custom Portfolio: Calculated weighted portfolio returns
weights = [1/3, 1/3, 1/3]

# "dreturns" = daily returns

combined_custom_dreturns_df = combined_custom_df.pct_change()

# "wreturns" = weighted returns

combined_custom_wreturns_df = combined_custom_dreturns_df.dot(weights)

combined_custom_wreturns_df.sort_index(ascending=True)
```

```python
# Custom Portfolio: Joined custom portfolio returns to the DataFrame that contains all of the portfolio returns

finalized_combined_df = pd.concat([combined_custom_wreturns_df, combined_df], axis = 'columns', join = 'inner')

finalized_combined_df.rename(columns = {0:'SHOP/OTEX/LOBLAW'}, inplace = True)

# Custom Portfolio: Compared dates where return data exists for all the stocks (drop NaNs)

finalized_combined_df.dropna(inplace = True)

finalized_combined_df.isnull().sum()

finalized_combined_df.sort_index(ascending=True)
```

```python
# Custom Portfolio: Calculated the Annualized Standard Deviation

finalized_combined_std = finalized_combined_df.std()*np.sqrt(252)

finalized_combined_std.sort_values(ascending= True, inplace=True)

finalized_combined_std
```

```python
# Custom Portfolio: Calculated and plotted rolling std with 21-day window 

finalized_combined_rolling_df = finalized_combined_df.rolling(window=21).std().plot(figsize = (20,10))
```

```python
# Custom Portfolio: Calculated and plotted the correlation

import seaborn as sns

custom_correlation_df = finalized_combined_df.corr()

custom_correlation_df

sns.heatmap(custom_correlation_df, vmin = -1, vmax = 1, annot=True, cmap="BuPu") 
```

```python
# Custom Portfolio: Calculated and Plotted the 60-day Rolling Beta for Custom Portfolio compared to the S&P 60 TSX

finalized_covariance = finalized_combined_df["SHOP/OTEX/LOBLAW"].rolling(window = 60).cov(finalized_combined_df["S&P TSX 60"])

finalized_variance = finalized_combined_df["S&P TSX 60"].rolling(window = 60).var()

finalized_combined_beta = finalized_covariance / finalized_variance

finalized_combined_beta.plot(figsize = (20,10))
```

```python
# Custom Portfolio: Calculated finalized sharpe ratio

finalized_sharpe_ratios = (finalized_combined_df.mean() * 252) / (finalized_combined_df.std() * np.sqrt(252))

finalized_sharpe_ratios
```

```python
# Custom Portfolio: Visualized the Sharpe ratios using bar plot

finalized_sharpe_ratios.plot(kind = "bar")
```
